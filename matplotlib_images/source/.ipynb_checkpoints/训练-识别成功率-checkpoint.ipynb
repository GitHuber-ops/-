```python
import matplotlib.pyplot as plt
import numpy as np

from pylab import mpl

mpl.rcParams['font.sans-serif'] = ['FangSong'] # 指定默认字体
mpl.rcParams['axes.unicode_minus'] = False # 解决保存图像是负号'-'显示为方块的问题


plt.rcParams['savefig.dpi'] = 300 #图片像素
plt.rcParams['figure.dpi'] = 300 #分辨率

x = np.linspace(0, 5000, 36)
print(x.size)

# Note that even in the OO-style, we use `.pyplot.figure` to create the figure.
fig, ax = plt.subplots()  # Create a figure and an axes.
ax.plot(x, raw_tr_acc, 'b',label='raw_tr')  # Plot some data on the axes.
ax.plot(x, improve_tr_acc,'r', label='improved_tr')  # Plot more data on the axes...

ax.set_xlabel('训练集大小')  # Add an x-label to the axes.
ax.set_ylabel('精确度')  # Add a y-label to the axes.
ax.set_title("训练集大小-企业实体识别成功率图")  # Add a title to the axes.
ax.legend()  # Add a legend.
plt.savefig("../images/训练集大小-识别成功率图.png")
```
